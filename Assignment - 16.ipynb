```python
try: 
    from PIL import Image
except ImportError:
    import Image
    
import pytesseract
import urllib.request
import cv2
import numpy as np
import imutils
import time
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def recText(filename):
    text = pytesseract.image_to_string(Image.open(filename))
    return text

# cam = cv2.VideoCapture(0)
# time.sleep(12)
url = 'http://192.168.1.7:8080/shot.jpg'

while True:
    imgPath = urllib.request.urlopen(url)
    imgNp = np.array(bytearray(imgPath.read()), dtype = np.uint8)
    
    # to take data and convert data into an image(imdecode)
    img = cv2.imdecode(imgNp, -1)
    img = imutils.resize(img, width = 450)
#     _, img = cam.read()
    cv2.imshow('Text Recognition', img)
    cv2.imwrite('imageFromCamera.png', img)
    key = cv2.waitKey(10)
    if key == 27: # pressing an escape button
        break

# cam.release()
cv2.destroyAllWindows()

info = recText('imageFromCamera.png')
print(info)
file = open('result.txt', 'w')
file.write(info)
file.close()
print('Successfully Written down ')
```
